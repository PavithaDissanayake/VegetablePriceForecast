```python
import pandas as pd
import os

folder = 'data/'
vegetables = sorted([f.split('.')[0] for f in os.listdir(folder) if f.endswith('.csv')][:-1])

dataframes = {}

for vegetable in vegetables:
    dataframes[vegetable] = pd.read_csv(f'{folder}/{vegetable}.csv', index_col='Date', parse_dates=True)

dataframes['Beans'].head()
```

```python
for vegetable in vegetables:
    for col in dataframes[vegetable].columns:
        dataframes[vegetable].rename(columns={col: f'{col}_{vegetable}'}, inplace=True)

combined_df = pd.concat(dataframes.values(), axis=1)
combined_df = combined_df[combined_df.index >= '2023-01-01']

combined_df.head()
```

```python
combined_df.to_csv('./Data/ultimate.csv')
```

```python
data = pd.read_csv('./Data/ultimate.csv')
data['Date'] = pd.to_datetime(data['Date'])
data.head(20)
```

```python
import firebase_admin
from firebase_admin import credentials, firestore
```

```python
for _, row in data.iterrows():
    print(str(row['Date'].date()))
```

```python
cred = credentials.Certificate('./firebase.json')
firebase_admin.initialize_app(cred)

db = firestore.client()

for _, row in data.iterrows():
    
    doc_id = str(row['Date'].date())
    row_dict = row.to_dict()

    db.collection('Data').document(doc_id).set(row_dict)
```
