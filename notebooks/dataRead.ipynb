```python
import pandas as pd

df = pd.read_csv('./Data/ultimate.csv')
df.head()
```

```python
import os

vegetables = sorted([f.split('.')[0] for f in os.listdir('./Data') if f.endswith('.csv')][:-1])
vegetables
```

```python
dataframes = {}
for vegetable in vegetables:
    markets = [col for col in df.columns if vegetable in col]
    data = pd.read_csv(f'./Data/{vegetable}.csv')
    temp = df[['Date'] + markets].copy()
    for market in markets:
        temp.rename(columns={market: market.split('_')[0]}, inplace=True)

    dataframes[vegetable] = pd.concat([data, temp], axis=0)

dataframes['Beans'].tail(20)
```
